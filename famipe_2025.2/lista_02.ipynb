## Exercício 01

**Herança e Polimorfismo** 
Qual será a saída do código abaixo?

```python
class Animal:
    def falar(self):
        return "Som do animal"

class Gato(Animal):
    def falar(self):
        return "Miau"

class Cachorro(Animal):
    pass # Não sobrescreve o método

animais = [Gato(), Cachorro()]
for animal in animais:
    print(animal.falar(), end=" ")
```

## Exercício 09

```python
class Conta:
    def __init__(self, titular):
        self.titular = titular
        self.saldo = 0

    def depositar(self, valor):
        self.saldo = self.saldo + valor

    def sacar(self, valor):
        self.saldo = self.saldo - valor

c = Conta("Maria")
c.depositar(100)
c.sacar(30)
c.depositar(50)
print(c.saldo)
```

## Exercício 10

```python
class Funcionario:
    def __init__(self, nome):
        self.nome = nome
        self.salario_base = 3000

    def calcular_salario(self):
        return self.salario_base

class Gerente(Funcionario):
    def __init__(self, nome):
        super().__init__(nome) # Chama o construtor da classe mãe
        self.bonus = 1500

    def calcular_salario(self): # Sobrescreve o método da classe mãe
        return self.salario_base + self.bonus

ger = Gerente("Carlos")
print(ger.calcular_salario())
```

```python
class Animal:
    def __init__(self, nome, especie):
        self.nome = nome
        self.especie = especie

class Ave(Animal):
    def __init__(self, nome, especie, envergadura_asas):
        super().__init__(nome, especie)
        self.envergadura_asas = envergadura_asas

    def exibir_dados(self):
        print(f"Animal: {self.nome} | Espécie: {self.especie} | Asas: {self.envergadura_asas}m")

# Teste
gaviao = Ave("Gavião Real", "Ave de Rapina", 2.0)
gaviao.exibir_dados()
```
